```python
def alpha_jn(j,n):
    if j>0 and j<n:
        alpha = (4/3)*((j-1)**(3/2)+(j+1)**(3/2)-2*j**(3/2))
    elif j==0:
        alpha = 4/3
    else:
        alpha = (4/3)*((n-1)**(3/2)-n**(3/2)+(6/4)*np.sqrt(n))
    return alpha
```

```python
def beta_jn(j,n):
    if n == 2:
        if j == 0:
            beta = (12/15)*np.sqrt(2)
        elif j == 1:
            beta = (16/15)*np.sqrt(2)
        else:
            beta = (2/15)*np.sqrt(2)
    elif n == 3:
        if j == 0:
            beta = (4/5)*np.sqrt(2)
        elif j == 1:
            beta = (14/5)*np.sqrt(3) - (12/5)*np.sqrt(2)
        elif j == 2:
            beta = -(8/5)*np.sqrt(3)+(12/5)*np.sqrt(2)
        else:
            beta = (4/5)*np.sqrt(3) - (4/5)*np.sqrt(2)
    else:
        if j>2 and j<n-1:
            beta = (8/15)*((j+2)**(5/2)-3*(j+1)**(5/2)+3*j**(5/2)-(j-1)**(5/2))
            +(2/3)*(-(j+2)**(3/2)+3*(j+1)**(3/2)-3*j**(3/2)+(j-1)**(3/2))
        elif j == 0:
            beta = (4/5)*np.sqrt(2)
        elif j == 1:
            beta = (14/5)*np.sqrt(3) - (12/5)*np.sqrt(2)
        elif j == 2:
            beta = (176/15) - (42/5)*np.sqrt(3) + (12/5)*np.sqrt(2)
        elif j == n-1:
            beta = (8/15)*(-2*n**(5/2)+3*(n-1)**(5/2)-(n-2)**(5/2))+(2/3)*(4*n**(3/2)-3*(n-1)**(3/2)+(n-2)**(3/2))
        else:
            beta = (8/15)*(n**(5/2)-(n-1)**(5/2))+(2/3)*(-3*n**(3/2)+(n-1)**(3/2))+2*np.sqrt(n)
    #print(beta)
    return beta
```

```python
def gamma_jn(j,n):
    if n == 3:
        if j == 0:
            gamma = (68/105)*np.sqrt(3)
        elif j == 1:
            gamma = (6/7)*np.sqrt(3)
        elif j == 2:
            gamma = (12/35)*np.sqrt(3)
        else:
            gamma = (16/105)*np.sqrt(3)
    elif n == 4:
        if j == 0:
            gamma = (244/315)*np.sqrt(2)
        elif j == 1:
            gamma = (1888/315) - (976/315)*np.sqrt(2)
        elif j == 2:
            gamma = -(656/105)+(488/105)*np.sqrt(2)
        elif j == 3:
            gamma = (544/105)-(976/315)*np.sqrt(2)
        else:
            gamma = -(292/315)+(244/315)*np.sqrt(2)
    elif n == 5:
        if j == 0:
            gamma = (244/315)*np.sqrt(2)
        elif j == 1:
            gamma = (362/105)*np.sqrt(3) - (976/315)*np.sqrt(2)
        elif j == 2:
            gamma = (500/63)*np.sqrt(5) - (1448/105)*np.sqrt(3) + (488/105)*np.sqrt(2)
        elif j == 3:
            gamma = -(290/21)*np.sqrt(5) + (724/35)*np.sqrt(3) - (976/315)*np.sqrt(2)
        elif j == 4:
            gamma = (220/21)*np.sqrt(5) - (1448/105)*np.sqrt(3) + (244/315)*np.sqrt(2)
        else:
            gamma = -(164/63)*np.sqrt(5) + (362/105)*np.sqrt(3)
    elif n == 6:
        if j == 0:
            gamma = (244/315)*np.sqrt(2)
        elif j == 1:
            gamma = (362/105)*np.sqrt(3) - (976/315)*np.sqrt(2)
        elif j == 2:
            gamma = (5584/315) - (1448/105)*np.sqrt(3) + (488/105)*np.sqrt(2)
        elif j == 3:
            gamma = (344/21)*np.sqrt(6) - (22336/315) + (724/35)*np.sqrt(3) - (976/315)*np.sqrt(2)
        elif j == 4:
            gamma = -(1188/35)*np.sqrt(6) + (11168/105) - (1448/105)*np.sqrt(3) + (244/315)*np.sqrt(2)
        elif j == 5:
            gamma = (936/35)*np.sqrt(6) - (22336/315) + (362/105)*np.sqrt(3)
        else:
            gamma = -(754/105)*np.sqrt(6) + (5584/315)
    else:
        if j>3 and j<n-3:
            gamma = (16/105)*((j+2)**(7/2)+(j-2)**(7/2)-4*(j+1)**(7/2)-4*(j-1)**(7/2)+6*j**(7/2))
            +(2/9)*(4*(j+1)**(3/2)+4*(j-1)**(3/2)-(j+2)**(3/2)-(j-2)**(3/2)-6*j**(3/2))
        elif j == 0:
            gamma = (244/315)*np.sqrt(2)
        elif j == 1:
            gamma = (362/105)*np.sqrt(3) - (976/315)*np.sqrt(2)
        elif j == 2:
            gamma = (5584/315) - (1448/105)*np.sqrt(3) + (488/105)*np.sqrt(2)
        elif j == 3:
            gamma = (1130/63)*np.sqrt(5) - (22336/315) + (724/35)*np.sqrt(3) - (976/315)*np.sqrt(2)
        elif j == n-3:
            gamma = (16/105)*(n**(7/2)-4*(n-2)**(7/2)+6*(n-3)**(7/2)-4*(n-4)**(7/2)+(n-5)**(7/2)) - 8/15*(n**(5/2))
            +(4/9)*(n**(3/2))+(8/9)*((n-2)**(3/2))-(4/3)*((n-3)**(3/2))+(8/9)*((n-4)**(3/2))-(2/9)*((n-5)**(3/2))
        elif j == n-2:
            gamma = (16/105)*((n-4)**(7/2)-4*(n-3)**(7/2)+6*(n-2)**(7/2)-3*(n**(7/2)))+(32/15)*(n**(5/2))-2*n**(3/2)
            -(4/3)*(n-2)**(3/2)+(8/9)*(n-3)**(3/2)-(2/9)*(n-4)**(3/2)
        elif j == n-1:
            gamma = (16/105)*(3*(n**(7/2))-4*(n-2)**(7/2)+(n-3)**(7/2)) - (8/3)*(n**(5/2)) + 4*n**(3/2) + (8/9)*(n-2)**(3/2)
            -(2/9)*(n-3)**(3/2)
        else:
            gamma = (16/105)*((n-2)**(7/2) - n**(7/2))+(16/15)*(n**(5/2))-(22/9)*(n**(3/2))-(2/9)*((n-2)**(3/2))+2*np.sqrt(n)
    return gamma
```

#### $\textbf{u} = |r|\hat{e}_{\phi}\Rightarrow u_x = -y,\;u_y = x$

```python
def velocity_u(r,t):
    x,y = r
    return -y,x

def Xi(R,S,h):
    return R*np.sqrt((3*h)/((np.pi)*S))

def G(r,t,u,w):
    x,y = r
    r_abs = np.sqrt(x**2+y**2)
    #phi = phil(y,r_abs)
    #print(phi)
    partial_duxdt = 0
    partial_duxdx = 0
    partial_duxdy = -1
    
    partial_duydt = 0
    partial_duydx = 1
    partial_duydy = 0
    v = w+u
    vx, vy = v
    exact_duxdt = partial_duxdt+vx*partial_duxdx+vy*partial_duxdy
    exact_duydt = partial_duydt+vx*partial_duydx+vy*partial_duydy

    Gx = (R-1)*exact_duxdt - R*np.dot(w,np.array([partial_duxdx, partial_duxdy])) - (R/S)*(w[0])
    Gy = (R-1)*exact_duydt - R*np.dot(w,np.array([partial_duydx, partial_duydy])) - (R/S)*(w[1])
    return Gx,Gy
```

```python
def First_order_sol(r0x,r0y,w0x,w0y,h,N):
    n = 0
    rx_sol = [r0x]
    wx_sol = [w0x]
    ry_sol = [r0y]
    wy_sol = [w0y]
    Gx = []
    Gy = []
    t = [0]
    while n<N:
        t0 = (n)*h
        t.append(t0+h)
        #print("t = ",t0)
        j = np.arange(0,n+1,1)
        u0x,u0y = velocity_u(np.array([r0x,r0y]),t0)
        G0x,G0y = G(np.array([r0x,r0y]),t0,np.array([u0x,u0y]),np.array([w0x,w0y]))
        #print(u0x,u0y)
        rx = r0x + h*(w0x+u0x)
        ry = r0y + h*(w0y+u0y)
        alpha_arr_1 = np.zeros(len(j))
        alpha_arr_2 = np.zeros(len(j))
        for ii in range(len(j)):
            alpha_arr_1[ii] = alpha_jn(j[ii]+1,n+1)
            alpha_arr_2[ii] = alpha_jn(j[ii],n)
        wx = (1/(1+xi*alpha_jn(0,n+1)))*(w0x+h*G0x - xi*(np.sum(np.dot(alpha_arr_1,np.flip(np.array(wx_sol))))
                                                  -np.sum(np.dot(alpha_arr_2,np.flip(np.array(wx_sol))))))
        wy = (1/(1+xi*alpha_jn(0,n+1)))*(w0y+h*G0y - xi*(np.sum(np.dot(alpha_arr_1,np.flip(np.array(wy_sol))))
                                                  -np.sum(np.dot(alpha_arr_2,np.flip(np.array(wy_sol))))))
        rx_sol.append(rx)
        wx_sol.append(wx)
        ry_sol.append(ry)
        wy_sol.append(wy)
        Gx.append(G0x)
        Gy.append(G0y)
        r0x = rx
        w0x = wx
        r0y = ry
        w0y = wy
        n = n+1
    return t, np.array(rx_sol),np.array(wx_sol),np.array(ry_sol),np.array(wy_sol),np.array(Gx), np.array(Gy)
```

```python
def Second_order_sol(rx_arr,ry_arr,wx_arr,wy_arr,h,N):
    n = 1
    r0x, r1x = rx_arr
    w0x, w1x = wx_arr
    r0y, r1y = ry_arr
    w0y, w1y = wy_arr
    rx_sol = [r0x, r1x]
    wx_sol = [w0x, w1x]
    ry_sol = [r0y, r1y]
    wy_sol = [w0y, w1y]
    Gx = []
    Gy = []
    t = [0,h]
    while n<N:
        t0 = (n)*h
        t.append(t0+h)
        j = np.arange(0,n+1,1)
        u1x,u1y = velocity_u(np.array([r1x,r1y]),t0)
        u0x,u0y = velocity_u(np.array([r0x,r0y]),t0-h)
        G1x,G1y = G(np.array([r1x,r1y]),t0,np.array([u1x,u1y]),np.array([w1x,w1y]))
        G0x,G0y = G(np.array([r0x,r0y]),t0-h,np.array([u0x,u0y]),np.array([w0x,w0y]))
        rx = r1x + (h/2)*(3*(w1x+u1x) - (w0x+u0x))
        ry = r1y + (h/2)*(3*(w1y+u1y) - (w0y+u0y))
        beta_arr_1 = np.zeros(len(j))
        beta_arr_2 = np.zeros(len(j))
        #print(n)
        #print(j)
        for ii in range(len(j)):
            #print(j[ii]+1,n+1)
            #print(j[ii],n)
            beta_arr_1[ii] = beta_jn(j[ii]+1,n+1)
            beta_arr_2[ii] = beta_jn(j[ii],n)
        wx = (1/(1+xi*beta_jn(0,n+1)))*(w1x+(h/2)*(3*G1x-G0x) - xi*(np.sum(np.dot(beta_arr_1,np.flip(np.array(wx_sol))))
                                                  -np.sum(np.dot(beta_arr_2,np.flip(np.array(wx_sol))))))
        wy = (1/(1+xi*beta_jn(0,n+1)))*(w1y+(h/2)*(3*G1y-G0y) - xi*(np.sum(np.dot(beta_arr_1,np.flip(np.array(wy_sol))))
                                                  -np.sum(np.dot(beta_arr_2,np.flip(np.array(wy_sol))))))
        rx_sol.append(rx)
        wx_sol.append(wx)
        ry_sol.append(ry)
        wy_sol.append(wy)
        Gx.append(G0x)
        Gy.append(G0y)
        r1x = rx
        w1x = wx
        r1y = ry
        w1y = wy
        r0x = rx_sol[-2]
        w0x = wx_sol[-2]
        r0y = ry_sol[-2]
        w0y = wy_sol[-2]
        n = n+1
    return t, np.array(rx_sol),np.array(wx_sol),np.array(ry_sol),np.array(wy_sol),np.array(Gx), np.array(Gy)
```

```python
def Third_order_sol(rx_arr,ry_arr,wx_arr,wy_arr,h,N):
    n = 2
    r0x, r1x, r2x = rx_arr
    w0x, w1x, w2x = wx_arr
    r0y, r1y, r2y = ry_arr
    w0y, w1y, w2y = wy_arr
    rx_sol = [r0x, r1x, r2x]
    wx_sol = [w0x, w1x, w2x]
    ry_sol = [r0y, r1y, r2y]
    wy_sol = [w0y, w1y, w2y]
    Gx = []
    Gy = []
    t = [0,h,2*h]
    while n<N:
        t0 = (n)*h
        t.append(t0+h)
        j = np.arange(0,n+1,1)
        u2x,u2y = velocity_u(np.array([r2x,r2y]),t0)
        u1x,u1y = velocity_u(np.array([r1x,r1y]),t0-h)
        u0x,u0y = velocity_u(np.array([r0x,r0y]),t0-2*h)
        G2x,G2y = G(np.array([r2x,r2y]),t0,np.array([u2x,u2y]),np.array([w2x,w2y]))
        G1x,G1y = G(np.array([r1x,r1y]),t0-h,np.array([u1x,u1y]),np.array([w1x,w1y]))
        G0x,G0y = G(np.array([r0x,r0y]),t0-2*h,np.array([u0x,u0y]),np.array([w0x,w0y]))
        rx = r2x + (h/12)*(23*(w2x+u2x) - 16*(w1x+u1x) + 5*(w0x+u0x))
        ry = r2y + (h/12)*(23*(w2y+u2y) - 16*(w1y+u1y) + 5*(w0y+u0y))
        gamma_arr_1 = np.zeros(len(j))
        gamma_arr_2 = np.zeros(len(j))
        for ii in range(len(j)):
            gamma_arr_1[ii] = gamma_jn(j[ii]+1,n+1)
            gamma_arr_2[ii] = gamma_jn(j[ii],n)
        wx = (1/(1+xi*gamma_jn(0,n+1)))*(w2x+(h/12)*(23*G2x - 16*G1x + 5*G0x) - xi*(np.sum(np.dot(gamma_arr_1,np.flip(np.array(wx_sol))))
                                                  -np.sum(np.dot(gamma_arr_2,np.flip(np.array(wx_sol))))))
        wy = (1/(1+xi*gamma_jn(0,n+1)))*(w2y+(h/12)*(23*G2y - 16*G1y + 5*G0y) - xi*(np.sum(np.dot(gamma_arr_1,np.flip(np.array(wy_sol))))
                                                  -np.sum(np.dot(gamma_arr_2,np.flip(np.array(wy_sol))))))
        rx_sol.append(rx)
        wx_sol.append(wx)
        ry_sol.append(ry)
        wy_sol.append(wy)
        Gx.append(G0x)
        Gy.append(G0y)
        r2x = rx
        w2x = wx
        r2y = ry
        w2y = wy
        r1x = rx_sol[-2]
        w1x = wx_sol[-2]
        r1y = ry_sol[-2]
        w1y = wy_sol[-2]
        r0x = rx_sol[-3]
        w0x = wx_sol[-3]
        r0y = ry_sol[-3]
        w0y = wy_sol[-3]
        n = n+1
    return t, np.array(rx_sol),np.array(wx_sol),np.array(ry_sol),np.array(wy_sol),np.array(Gx), np.array(Gy)
```

#### Solution for $\textbf{r}_0 = (1,0)$ and $\textbf{w}_0 = (0,0)$

```python
R = 0.75
S = 0.3
h = 0.01
xi = Xi(R,S,h)
N = 2000
t,rx,wx,ry,wy,gx,gy = First_order_sol(1,0,0,0,h,N)
```

```python
R = 0.75
S = 0.3
h = 0.01
xi = Xi(R,S,h)
N = 2000
t2,rx2,wx2,ry2,wy2,gx2,gy2 = Second_order_sol([1,rx[1]],[0,ry[1]],[0,wx[1]],[0,wy[1]],h,N)
```

```python
plt.figure(figsize = (6,5.5))
plt.plot(t,rx,label = "x")
plt.plot(t,ry,label = "y")
plt.legend()
```

```python
Cs = 0
gamma = 1.5
S = (0.3)/(3)
lambda_r = 2*S*(gamma-1)-3*np.sqrt(2)*(S**(3/2))*(gamma-1)
lambda_i = 1+Cs*np.sqrt(2)*(S**(3/2))*(gamma-1)/np.pi-3*np.sqrt(2)*(S**(3/2))*(gamma-1)
lamda = (lambda_r+1j*lambda_i)
t = np.linspace(0,20.85,2001)   
## Here,final value of time is a bit more than the final value we got for 1st and 2nd order solution. It is because of the fact that 
## the analytic solution is ~exp(lamda*t) and it works well for large time. For small time, solution may differ at particular time point
## but the trajectory for large time would remain the same. This analytic solution is just an approximation but it can be 
## roughly used to show the trajectory of particle even at smaller times. 
Zt = np.exp(lamda*np.array(t))
X,Y = np.real(Zt),np.imag(Zt)
```
